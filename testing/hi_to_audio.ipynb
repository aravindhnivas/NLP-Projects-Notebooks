```python
from transformers import VitsModel, AutoTokenizer
import torch
from IPython.display import Audio
import scipy
import numpy as np
```

```python
model = VitsModel.from_pretrained("facebook/mms-tts-hin")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-hin")

text = "सन्‌ 1813 में प्रकाशित हुए अँग्रेज़ लेखक जेन ए.ए. ए. के मुताबिक, यह शिष्टाचार का दूसरा संक्षिप्त रूप है, यह एल.टी. में दी गई पुस्तक के उपादेशीय विकास के अनुरूप है, जो तुरंत न्याय के बारे में सीखते हैं और ऊपरी तथा वास्तविक भलाई के बीच अंतर की क़दर करने लगता है।"
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform
```

```python
print("Sampling rate:", model.config.sampling_rate)
print("Number of channels:", output.float().numpy().shape[0])
```

```python
output.float().numpy()
```

```python
# Convert the output to int16
output_int16 = (output.numpy() * 32767).astype(np.int16)

# Write the output to a WAV file
scipy.io.wavfile.write("output.wav", model.config.sampling_rate, output_int16)
```
